Here imputing values for the MitreTechniques column and encoding the 'MitreTechniques' column using the MultiLabelBinarizer.

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Step 1: Impute missing values in MitreTechniques with 'Unknown'
new_train_sample['MitreTechniques'] = new_train_sample['MitreTechniques'].fillna('Unknown')

# Step 2: Apply binary encoding to the 'MitreTechniques' column
# First, split the techniques by semicolon and then use MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Split the techniques by semicolon and apply binary encoding
technique_dummies = mlb.fit_transform(new_train_sample['MitreTechniques'].apply(lambda x: x.split(';')))

# Step 3: Create a DataFrame with binary columns for each technique
technique_df = pd.DataFrame(technique_dummies, columns=mlb.classes_, index=new_train_sample.index)

# Step 4: Join the binary technique columns back to the original dataset
new_train_sample = new_train_sample.join(technique_df)

# Now, new_train_sample contains the binary encoded columns for MitreTechniques, and the missing values have been imputed
print("Data after imputing and encoding:")
print(new_train_sample.head())

# Optionally, you can drop the original 'MitreTechniques' column after encoding
new_train_sample = new_train_sample.drop(columns=['MitreTechniques'])

# Check the final dataset
print("Final Data after dropping 'MitreTechniques':")
print(new_train_sample.head())
```

```python
import numpy as np
def handle_outliers_iqr(df, column_name):
    """
    Handle outliers in a specified column using the IQR method.
    """
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Determine lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap outliers
    df[column_name] = np.clip(df[column_name], lower_bound, upper_bound)
    
    return df

# Apply IQR outlier handling to the 'age' column
new_train_sample = handle_outliers_iqr(new_train_sample, 'hour')

print("Data shape after IQR handling for 'hour':", new_train_sample.shape)
```

```python
from scipy.stats import skew
from sklearn.preprocessing import PowerTransformer
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Select the column to transform
column = new_train_sample['hour']

# Step 2: Reshape the data
column_reshaped = column.values.reshape(-1, 1)

# Step 3: Apply Yeo-Johnson Transformation
pt = PowerTransformer(method='yeo-johnson')
transformed_column = pt.fit_transform(column_reshaped)

# Step 4 (Optional): Check the statistics after transformation (but do not save it to the DataFrame)
print(f"Original skewness: {np.round(skew(column), 4)}")
print(f"Transformed skewness: {np.round(skew(transformed_column.flatten()), 4)}")

# Step 5: Visualize the change

# Flatten the transformed data for plotting
transformed_column_flat = transformed_column.flatten()

# Setting up the figure size
plt.rcParams['figure.figsize'] = (10, 5)

# Creating subplots
fig, ax = plt.subplots(1, 2)

# Original data distribution
sns.distplot(column, label="Original skew: {0}".format(np.round(skew(column), 4)),
             color='r', ax=ax[0], axlabel='Original')

# Yeo-Johnson transformed data distribution
sns.distplot(transformed_column_flat, label="Transformed skew: {0}".format(np.round(skew(transformed_column_flat), 4)),
             color='g', ax=ax[1], axlabel='Yeo-Johnson Transform')

# Add legend
fig.legend()
plt.show()
```

```python
import numpy as np
def handle_outliers_iqr(df, column_name):
    """
    Handle outliers in a specified column using the IQR method.
    """
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Determine lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap outliers
    df[column_name] = np.clip(df[column_name], lower_bound, upper_bound)
    
    return df

# Apply IQR outlier handling to the 'age' column
new_train_sample = handle_outliers_iqr(new_train_sample, 'AlertId')

print("Data shape after IQR handling for 'AlertId':", new_train_sample.shape)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming your final resampled data is X_resampled and y_resampled
# Step 1: Split the Data
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

# Step 2: Standardize the Data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Logistic Regression
print("Training Logistic Regression...")
logistic_model = LogisticRegression(multi_class='ovr', max_iter=1000, random_state=42)
logistic_model.fit(X_train, y_train)

# Evaluate Logistic Regression
y_pred = logistic_model.predict(X_test)
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest
print("Training Random Forest...")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Evaluate Random Forest
y_pred = rf_model.predict(X_test)
print(f"Random Forest Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Confusion Matrix - Random Forest")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
